# 데이터 로드

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
```

# 데이터 준비 및 탐색

```python
# 데이터 설명 확인
print(data.DESCR)
```

**Attribute Information**
- radius : 반지름(중심에서 둘레의 점까지 거리의 평균)
- texture : 텍스처(그레이 스케일 값의 표준편차)
- perimeter : 둘레
- area : 영역
- smoothness : 평활도(반지름 길이의 국부적 변동)
- compactness : 콤팩트함(둘레^2/면적 - 1.0)
- concavity : 오목(윤곽선의 오목한 부분의 심각도)
- concave points : 오목점(윤곽선의 오목한 부분의 수)
- symmetry : 대칭
- fractal dimension : 프랙탈 차원("해안선 근사치" - 1)

# 데이터 전처리

```python
# 타겟 데이터 저장
target = data_df['classification']
data_df.drop(labels=['classification'], axis=1, inplace=True)
data_df
```

```python
# Standard Scaler로 data scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(data_df)
data_prepare = scaler.transform(data_df)
data_prepare
```

# 분석 모델 구축

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# 훈련용 데이터와 평가용 데이터 분할
feature_train, feature_test, target_train, target_test = train_test_split(data_prepare, target, test_size=0.2, random_state=42)
```

```python
# 로지스틱 회귀 분석 : (1) 모델 생성
lr_model = LogisticRegression()
```

```python
# 로지스틱 회귀 분석 : (2) 모델 훈련
lr_model.fit(feature_train, target_train)
```

```python
# 로지스틱 회귀 분석 : (3) 평가 데이터에 대한 예측 수행 -> 예측 결과 Y_predict 구하기
target_predict = lr_model.predict(feature_test)
print(target_predict)
```

# 결과 분석

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
```

```python
# 오차 행렬
confusion_matrix(target_test, target_predict)
```

```python
acccuracy = accuracy_score(target_test, target_predict)
precision = precision_score(target_test, target_predict)
recall = recall_score(target_test, target_predict)
f1 = f1_score(target_test, target_predict)
roc_auc = roc_auc_score(target_test, target_predict)
```

```python
print('정확도: {0:.3f}, 정밀도: {1:.3f}, 재현율: {2:.3f},  F1: {3:.3f}'.format(acccuracy,precision,recall,f1))
```

```python
print('ROC_AUC: {0:.3f}'.format(roc_auc))
```
